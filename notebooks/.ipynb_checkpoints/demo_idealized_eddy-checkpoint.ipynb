# Notebook: test algo on idealized eddy shape 

This notebook aims to document the behaviour of the geostrophic current and vorticity calculation algorithm from an idealised SSH field. An idealized vortex shape (on which the theoretical solutions of the geostrophic and vorticity currents can be computed) is parameterized as input. The solution calculated by the geostrophic and vorticity algorithm can then be compared with the theoretical solution.

### Idealized Eddy properties

```python
eddy_properties = {
    'eddy_center_lat' : 40.,         # degree north
    'eddy_center_lon' : 0.,          # degree east
    'eddy_radius_rx' : 100000.,      # eddy x radius in meters 
    'eddy_radius_ry' : 100000,       # eddy y radius in meters 
    'eddy_amp' : 0.1,                # eddy mplitude radius in meters 
    'dx' : 2000.,                    # x grid step in meters
    'dy' : 2000.                     # y grid step in meters
}
```

```python
ds_eddy = generate_gaussian_eddy_shape(eddy_properties)
```

```python
plot_eddy(ds_eddy)
```

```python
params = {
    'dx' : eddy_properties['dx'],             # SWOT swath x-grid step in meters
    'dy' : eddy_properties['dy'],             # SWOT swath y-grid step in meters
    'stencil_npt' : 3,                       # stencil width: number of point for discretization
    'filtrage_m' : 300000,                    # lenght scale (in meters) equatorial filtering
    'lat_max_equator' : 10.,                  # maximum latitude (in degree) to consider for equatorial band processing (Recomm: not less than 10)
    'maxval_current' : 10.                    # (absolute) maximum value of surface currents in meter / second
}
```

```python
u_g, v_g = compute_karin_geos_current(ds_eddy, params)
zeta_f = compute_vorticity(u_g, v_g, params)
```

```python
(numpy.sqrt(u_g**2 + v_g**2) - numpy.sqrt(ds_eddy.ug_theory**2 + ds_eddy.vg_theory**2)).plot(vmin=-0.0000001, vmax=0.0000001, cmap='coolwarm')
```

```python
(100*(v_g - ds_eddy.vg_theory)/ds_eddy.vg_theory).plot(vmin=-0.1, vmax=0.1, cmap='coolwarm')
```

```python
(100*(u_g - ds_eddy.ug_theory)/ds_eddy.ug_theory).plot(vmin=-0.1, vmax=0.1, cmap='coolwarm')
```

```python
(100*(zeta_f - ds_eddy.zeta_f)/ds_eddy.zeta_f).plot(vmin=-1, vmax=1, cmap='coolwarm')
```
